# Step 1: Data reading and inspecting

```python
# Supress Warnings

import warnings
warnings.filterwarnings('ignore')
```

```python
# Import the numpy and pandas package

import numpy as np
import pandas as pd
```

```python
# Read the given CSV file, and view some sample records

df = pd.read_csv("day.csv")
df.head()
```

```python
df.shape
```

```python
# Drop 'instant' and dteday column as 'instant' does not serve any purpose and dteday column vital informations have been extracted
df.drop(['instant'],axis=1,inplace=True)
df.drop(['dteday'],axis=1,inplace=True)
```

```python
df.info()
```

```python
df.describe()
```

# Step 2: Data Conversion

```python

# Replace season binary numbers into meaningful categories. This would change the dtype to object too.
df['season'].replace({1:'spring',2:'summer',3:'fall',4:'winter'},inplace=True)
df.head()
```

```python
# Replace yr binary numbers into meaningful categories. This would change the dtype to object too.
df['yr'].replace({0: '2018', 1:'2019'},inplace=True)
df.head()
```

```python
df['mnth'].replace({1:'January', 2:'Febraury',3:'March',4:'April',5:'May',6:'June',7:'July',8:'August',9:'September',10:'October',11:'Novmber',12:'December'},inplace=True)
df.head()
```

```python
# Change column 'holiday' into meaninfgul names. This would change the dtype to object too.
df['holiday'].replace({0:'No Holiday',1:'Holiday'},inplace=True)
df.head()
```

```python
# Change the weekday column encoding to meaningful categories. This would change the dtype to object too.
df['weekday'].replace({0:'Mon',1:'Tue',2:'Wed',3:'Thurs',4:'Fri',5:'Sat',6:'Sun'},inplace=True)
df.head()
```

```python
# Change column 'workingday' sub-variable into more meaninfgul name. This would change the dtype to object too.
df['workingday'].replace({0:'workingday',1:'Holiday_w'},inplace=True)
df.head()
```

```python
# Change the weekday column encoding to meaningful categories. This would change the dtype to object too.
df['weathersit'].replace({1:'Clear', 
                          2:'Cloudy',
                          3:'Light Rain',
                          4:'Heavy Rain'},inplace=True)
df.head()
```

# Step 4: Data Preparation

```python
#Create dummy variables for all categorical variables
#Print few rows to inspect

bike_df = pd.get_dummies(df[['yr','holiday','season','mnth','weekday','weathersit','workingday']],drop_first=True)
bike_df.head()
```

```python
# We then join orignal data and the dummy variable dataframe 'status'.
final=pd.concat([df,bike_df],axis=1)
final.head()
```

```python
# We then drop the indivisual columns whose dummy variables have been created.
final=final.drop(['season','yr','holiday','mnth','workingday','weekday','weathersit'],axis=1)
final.head()
```

# Step 5: Splitting the Data into Training and Testing Sets


```python

# Import libraries to train and test split
import sklearn
from sklearn.model_selection import train_test_split
```

```python
# We specify this so that the train and test data set always have the same rows, respectively
np.random.seed(0)
# Create df_train and df_test split from the dataframe and check their shape
df_train,df_test=train_test_split(final,train_size=0.7,random_state=100)
print(df_train.shape)
print(df_test.shape)
```

```python
# Check the parameters obtained

lr.params
```

```python
# Let's visualise the data with a scatter plot and the fitted regression line
plt.scatter(X_train_lm.iloc[:, 1], y_train)
plt.plot(X_train_lm.iloc[:, 1], 0.0243 + 0.933*X_train_lm.iloc[:, 1], 'r')
plt.show()
```

```python
# Print a summary of the linear regression model obtained
print(lr.summary())
```

# Checking VIF

```python
# Check for the VIF values of the feature variables. 
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# Create a dataframe that will contain the names of all the feature variables and their respective VIFs
# Let's check it's ViF
vif = pd.DataFrame()
X = X_train
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```
